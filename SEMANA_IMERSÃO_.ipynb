```python
# Desafio 5 - calcular e analizar a correlação entre G(gênica) e c(tipos celulares) plotando o gráfico
# DICA: vai estar plontando matriz quadrada pensar em faz um heatmap com eles
# Refletir sobre efeitos biologicos(Porque a correlação alta de alguns genes para alguns tipos celulares)
# Compute the correlation matrix
corr_genes = dados.loc[:,'g0':'g20']
corr_celulas = dados.loc[:,'c0':'c20']
genes_celulas = pd.concat([corr_genes,corr_celulas],axis=1)
corr = genes_celulas.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
dados_resultados['n_moa']= dados_resultados.drop('id', axis=1).sum(axis=1)
dados_resultados.head()
```

```python
#---1
# Juntar 2 bases de dados(para pegar informações(colunas ou linhas) entre base de dados)

# Cria nova coluna n moa em     dados_resultados.drop('id', axis=1).sum(axis=1)    criada anteriormente
# (Já executou lá em cima: dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)

# Cria coluna amostrando se foi ativa o moa (mecanismo de ação) 
#     dados_resultados['n_moa'] != 0    Se na coluna moa  1=True  e 0=False
dados_resultados['ativo_moa'] = dados_resultados['n_moa'] != 0
dados_resultados.head()
```
